```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

#dados = pd.read_csv("https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true")

# Armazenando os dados do arquivo .csv em um Dataframe chamado dados
dados = pd.read_csv("dataset/MICRODADOS_ENEM_2019.csv")
dados.head()
```

```python
# Selecionando apenas candidatos com 13 anos de idade
dados.query("NU_IDADE == 13")
```

```python
# Fazendo a contagem dos alunos com 14 anos de idade ou menos por estado
dados.query("NU_IDADE <= 14")["SG_UF_RESIDENCIA"].value_counts()
```

```python
# Fazendo a contagem dos alunos com 14 anos de idade ou menos por estado normalizado (porcentagem)
dados.query("NU_IDADE <= 14")["SG_UF_RESIDENCIA"].value_counts(normalize=True)
```

```python
# Plotando um gráfico de pizza dos alunos com 14 anos de idade ou menos por estado
alunos_menor_quartoze = dados.query("NU_IDADE <= 14")
plt.title("Distribuição dos Inscritos com Menos de 15 anos por Estado")
alunos_menor_quartoze["SG_UF_RESIDENCIA"].value_counts().plot.pie(figsize=(12,12));
```

```python
# Plotando um gráfico de barra dos alunos com 14 anos de idade ou menos por estado
plt.title("Distribuição dos Inscritos com Menos de 15 anos por Estado")
alunos_menor_quartoze["SG_UF_RESIDENCIA"].value_counts(normalize=True).plot.bar(figsize=(12,8));
```

```python
# Ordenando os valores presentes como resposta da questão Q0006 para [A, B, ..., P]
renda_ordenada = dados["Q006"].unique()
renda_ordenada.sort()

# Mostrando através de um boxplot a relação entre a renda e a nota de matemática
plt.figure(figsize=(12, 10))
plt.title("Notas de Matemática pela Renda")
sns.boxplot(data=dados, x="Q006", y = "NU_NOTA_MT", order=renda_ordenada);
```

```python
# Delimitando quais colunas possuem provas
provas = ["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_MT", "NU_NOTA_LC", "NU_NOTA_REDACAO"]

# Criando uma coluna no DataFrame contendo a soma das cinco provas
dados["NU_NOTA_TOTAL"] = dados[provas].sum(axis=1)
dados[["NU_NOTA_TOTAL", "NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_MT", "NU_NOTA_LC", "NU_NOTA_REDACAO"]].head()
```

```python
# Mostrando através de um boxplot a relação entre a renda e a nota de matemática
plt.figure(figsize=(12, 10))
plt.title("Nota Total pela Renda")
sns.boxplot(x="Q006", y = "NU_NOTA_TOTAL", data = dados, order = renda_ordenada);
```

```python
# Visualizando a distribuição das notas totais dos inscritos
sns.displot(dados, x="NU_NOTA_TOTAL");
```

```python
# Acrescentando a nota 'NU_NOTA_TOTAL' na lista de provas
provas.append("NU_NOTA_TOTAL")

dados[provas].query("NU_NOTA_TOTAL == 0")
```

```python
# Removendo os inscritos que possuem nota total 0
dados_sem_notas_zero = dados.query("NU_NOTA_TOTAL != 0")
dados_sem_notas_zero.head()
```

```python
# Mostrando através de um boxplot a relação entre a renda e a nota total (sem as notas 0)
plt.figure(figsize=(12, 10))
plt.title("Nota Total pela Renda")
sns.boxplot(x="Q006", y="NU_NOTA_TOTAL", data=dados_sem_notas_zero, order=renda_ordenada);
```

```python
plt.figure(figsize=(14, 8))
plt.title("Relação entre a Nota Total e a Renda")
sns.boxplot(x="Q006", y="NU_NOTA_TOTAL", data=dados_sem_notas_zero,
            hue="IN_TREINEIRO", order=renda_ordenada);
```

```python
# Desafio 1: Aumentar a amostra de alunos com mais de 14 anos para alunos menores de idade e comparar a proporção por estado

alunos_menor_idade = dados.query("NU_IDADE <= 18")
plt.title("Distribuição dos Inscritos Menores de Idade por Estado")
alunos_menor_idade["SG_UF_RESIDENCIA"].value_counts(normalize=True).plot.bar(figsize=(12,8));
```

```python
# Desafio 3: Criar uma função para plotar o boxplot no seaborn

def plotar_boxplot(data, x, y, cores, ordem):
    plt.figure(figsize=(14, 8))
    return sns.boxplot(x=x, y=y, data=data, hue=cores, order=ordem);

plotar_boxplot(data=dados_sem_notas_zero, x='Q006', y='NU_NOTA_TOTAL', cores='IN_TREINEIRO', ordem=renda_ordenada)
```

```python
# Desafio 4: Verificar se quem zerou a prova foi eliminado ou não estava presente.

#Agrupar dados de presença por materia
dados_zerados = ["TP_PRESENCA_CH", "TP_PRESENCA_LC", "TP_PRESENCA_CN", "TP_PRESENCA_MT", 'NU_NOTA_TOTAL']

presenca_zerados = dados[dados_zerados].query("NU_NOTA_TOTAL == 0")

presenca_zerados.drop(columns="NU_NOTA_TOTAL", inplace=True)

presenca_zerados.apply(pd.Series.value_counts)
```

```python
# Desafio 5: Quem é eliminado tira zero ou é preenchido com NaN (não teve registro de notas).

#Eliminados na prova CH
print(dados.query("TP_PRESENCA_CH == 2")["NU_NOTA_CH"].value_counts(dropna=False))

#Eliminados na prova LC
print(dados.query("TP_PRESENCA_LC == 2")["NU_NOTA_LC"].value_counts(dropna=False))

#Eliminados na prova CN
print(dados.query("TP_PRESENCA_CN == 2")["NU_NOTA_CN"].value_counts(dropna=False))

#Eliminados na prova MT
print(dados.query("TP_PRESENCA_MT == 2")["NU_NOTA_MT"].value_counts(dropna=False))
```

```python
#Desafio 7: Fazer um boxplot olhando para a questão 25 (tem internet ou não) e fazer uma reflexão sobre o assunto.
# A = Não possui Internet, B = Possui Internet

plt.figure(figsize=(14, 8))
plt.title("Relação entre a Nota Total e a Renda para Inscritos que Possuem e Não Possuem Internet")
sns.boxplot(x="Q006", y="NU_NOTA_TOTAL", data=dados_sem_notas_zero,
            hue="Q025", order=renda_ordenada);
```
